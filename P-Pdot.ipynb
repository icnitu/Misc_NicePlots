```python
from psrqpy import QueryATNF
```

```python
query = QueryATNF(params=['P0', 'P1'])
```

```python
fig = query.ppdot(showSNRs=True, showGCs=True, showtypes=['BINARY'], rcparams={'legend.frameon':True, \
                'xtick.direction':'in', 'ytick.direction':'in', 'ytick.major.size':7, 'xtick.major.size':7, \
                 'ytick.minor.size':4, 'xtick.minor.size':4, 'font.family': 'serif'},\
                markertypes={'GC':{'marker':'s', 'markeredgecolor': 'royalblue', 'markersize':10}, \
                             'SNR':{'marker':'*', 'markeredgecolor': 'purple','markersize':8},\
                             'BINARY':{'marker':'o', 'markeredgecolor': 'gray','markersize':6}},\
                tau=[1e2, 1e4, 1e6, 1e8, 1e10], Bfield=[1e8, 1e10, 1e12, 1e14], periodlims=[3e-4, 3e1],\
                pdotlims=[7e-22, 2e-9])
                                                                               
# fig'constrained_layout':True
# fig.tight_layout()
fig.savefig('PPdot_10Oct2022.pdf')
plt.show()
```

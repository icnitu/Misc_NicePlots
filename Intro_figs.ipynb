```python
def circle(c,r,th,rot):
    theta = np.linspace(0, th, 360) + rot
    x = r * np.cos(theta) + c[0]
    y = r * np.sin(theta) + c[1]
    return(x,y)

def ellipse(c,a,b,th,rot):
    theta = np.linspace(0, th, 360) + rot
    x = a * np.cos(theta) + c[0]
    y = b * np.sin(theta) + c[1]
    return(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def zigzag(x0, xf, lowy):
    
    xs = np.arange(2*x0, 2*xf+1)/2
    
    ys = np.zeros_like(xs)
    
    for i,x in enumerate(xs):
    
        if int(x) == x:

            ys[i] = lowy + 0.5*(int(lowy) == lowy)

        else:

            ys[i] = lowy + 0.5*(int(lowy) != lowy)
            
    
    return(xs, ys)
```

```python
nx = 10; ny=16
```

```python
def basic_midrect(col1, col2, nx=6, ny=8, figsizes=(3,5)):
    
    fig, ax = plt.subplots(1, 1, figsize=figsizes, constrained_layout=True)
    
    intx = np.arange(nx)
    inty = np.arange(ny)
    
    for x in range(2*nx): 
        ax.axvline(x/2, ls='--', c='xkcd:gray')
    for y in range(2*ny): 
        ax.axhline(y/2, ls='--', c='xkcd:gray')
    

    for y in np.arange(2*ny)/2:
        
        xs, ys = zigzag(0, nx, y)
        ax.plot(xs, ys, ls='-', c=col1)
        
    for y in np.arange(ny/2, 3/2*ny)/2:
        
        xs, ys = zigzag(nx/4, 3/4*nx, y)
        ax.plot(xs, ys, ls='-', c=col2)
    
    ax.fill_between([nx/4, 3/4*nx], ny/4, 3/4*ny, color=col2, alpha=0.5)
    
    plt.xlim(0, nx)
    plt.ylim(0, ny)
    
#     plt.show()
    return fig, ax
    
fig, ax = basic_midrect('xkcd:gold', 'xkcd:royal blue', nx=nx, ny=ny)
# fig.show()
```

```python
def farfarelle_midrect(col1, col2, nx=6, ny=8, figsizes=(3,5)):
    
    fig, ax = basic_midrect(col1, col2, nx, ny, figsizes)
    
    
    
    for y in [0.5, ny-1]:
        
        xs, ys = zigzag(0.5, nx-0.5, y)

        ax.plot(xs, ys, ls='-', c=col2)

        
    for dy in np.arange(1,ny/2)/2:

        for y in [dy+0.5, ny-1-dy]:

            xs, ys = zigzag(dy, dy+0.5, y)
            ax.plot(xs, ys, ls='-', c=col2)

            xs, ys = zigzag(nx-0.5-dy, nx-dy, y)
            ax.plot(xs, ys, ls='-', c=col2)

#     for y in [1.5, ny-2]:
#         xs, ys = zigzag(1, 1.5, y)
#         ax.plot(xs, ys, ls='-', c=col2)

#         xs, ys = zigzag(nx-1.5, nx-1, y)
#         ax.plot(xs, ys, ls='-', c=col2)
        
        

#     plt.show()

    return fig, ax
    
fig, ax = farfarelle_midrect('xkcd:gold', 'xkcd:royal blue', nx=nx, ny=ny)
```

```python
def bra_ket_midrect(col1, col2, nx=6, ny=8, figsizes=(3,5)):
    
    fig, ax = basic_midrect(col1, col2, nx, ny, figsizes)
    
    
    for y in [0.5, 1, ny-1.5, ny-1]:
        
        xs, ys = zigzag(1.5, nx-1.5, y)

        ax.plot(xs, ys, ls='-', c=col2)
        
    for y in [3/8*ny-0.5, 3/8*ny, 5/8*ny-0.5, 5/8*ny]:
        
        xs, ys = zigzag(0.5, 1, y)

        ax.plot(xs, ys, ls='-', c=col2)  


        xs, ys = zigzag(nx-1, nx-0.5, y)

        ax.plot(xs, ys, ls='-', c=col2)  

    
#     plt.show()

    return fig, ax
    
fig, ax = bra_ket_midrect('xkcd:gold', 'xkcd:royal blue', nx=nx, ny=ny)
```

```python
def frillyedge_midrect(col1, col2, nx=6, ny=8, figsizes=(3,5)):
    
    fig, ax = basic_midrect(col1, col2, nx, ny, figsizes)
    
    
    
    for y in [0.5, ny-1]:
        
        xs, ys = zigzag(0.5, nx-0.5, y)

        ax.plot(xs, ys, ls='-', c=col2)

    for y in np.arange(2, 2*ny-2)/2:

        xs, ys = zigzag(0.5, 1, y)
        ax.plot(xs, ys, ls='-', c=col2)

        xs, ys = zigzag(nx-1, nx-0.5, y)
        ax.plot(xs, ys, ls='-', c=col2)


        

#     plt.show()

    return fig, ax
    
fig, ax = frillyedge_midrect('xkcd:gold', 'xkcd:royal blue', nx=nx, ny=ny)
```
